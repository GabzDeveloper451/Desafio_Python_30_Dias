```python
#Crie um programa que peça um número, mas use try/except para lidar se o usuário digitar texto.

#definindo input e regras
while True:
  try:
    numero=int(input("Digite um número inteiro: "))
    print(f"você digitou {numero}")
    break
  except ValueError:
    print("tente novamente!")
```
