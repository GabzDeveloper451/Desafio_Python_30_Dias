```python
#Crie dois conjuntos de números e encontre a união, interseção e diferença entre eles.

#definindo conjuntos
conj1 = {1, 2, 3, 4, 5}
conj2 = {4, 5, 6, 7, 8}

print(f"Conjunto 1: {conj1}")
print(f"Conjunto 2: {conj2}")

#apresentando união
uniao= conj1.union(conj2)
print(f"União: {uniao}")
#apresentando intersecção
interseccao = conj1.intersection(conj2)
print(f"Intersecção: {interseccao}")
#apresentando diferenças
diferenca1 = conj1.difference(conj2)
diferenca2 = conj2.difference(conj1)
print(f"Diferença do conjunto 1 para o conjunto 2): {diferenca1}")
print(f"Diferença do conjunto 2 para o conjunto 1): {diferenca2}")
```
