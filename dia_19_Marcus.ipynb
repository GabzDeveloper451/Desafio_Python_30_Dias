```python
#Crie um dicionário para armazenar informações de uma pessoa (nome, idade, cidade).

#criando dicionário
dic_dados={"nome":"Marcus", "idade":"36", "cidade":"Itaquaquecetuba"}

#exibindo dicionário
print(dic_dados)
```

```python
#Adicione uma nova chave, atualize um valor e use .keys() e .values() para
#imprimir as chaves e valores separadamente.

# nova chave
dic_dados["profissao"] = "biólogo e analista de dados"

# Atualizando idade
dic_dados["idade"] = 37

# Imprimindo as chaves
print("Chaves do dicionário:", dic_dados.keys())

# Imprimindo os valores
print("Valores do dicionário:", dic_dados.values())

# Exibindo o dicionário atualizado
print("Dicionário atualizado:")
print(dic_dados)
```
