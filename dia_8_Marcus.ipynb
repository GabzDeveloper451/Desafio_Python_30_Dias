```python
#Adicione uma fruta, remova outra e ordene a lista em ordem alfabética.

lista_de_trufas=["jaranla","acabate", "bermogata", "nabana", "vua" ]

print(lista_de_trufas[4])
print(lista_de_trufas[3])
print(lista_de_trufas[2])
print(lista_de_trufas[1])
print(lista_de_trufas[0])

#removendo um item
print("lista com remoção")
lista_de_trufas.remove("nabana")
print(lista_de_trufas)

#adicionando item
print("lista com adição")
lista_de_trufas.append("banana")
print(lista_de_trufas)

#ordenando alfabeticamente
print("lista ordenada")
lista_de_trufas.sort()
print(lista_de_trufas)
```
