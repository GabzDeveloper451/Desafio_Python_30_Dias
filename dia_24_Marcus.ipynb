```python
#Crie um script que gere uma senha aleatória de 8 caracteres (letras e números).

#importando bibliotecas random para aleatoriedade e string para caracteres
import random
import string


#definindo critérios para criação da senha
def senha_aleatoria(legth=8):
    caracteres = string.ascii_letters + string.digits
    senha = ''.join(random.choice(caracteres) for i in range(legth))
    return senha

#criando senha
senha=senha_aleatoria(8)
print(f"Sua senha aleatória é: {senha}")
```
