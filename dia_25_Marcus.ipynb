```python
#Implemente um jogo onde o computador gera um número secreto e o usuário tenta adivinhar.

#importando biblioteca
import random as rd

#definindo número aleatório inteiro
num_sec=rd.randint(1,100)

#chamada para o desafio
print("pensei em um número de 1 a 100, vamos ver se você advinha")


#configuração do desafio
while True:
  try:
    palpite = int(input("palpite: "))
    if palpite == num_sec:
      print ("você acertou!!!!")
      break
    elif palpite > num_sec:
      print ("tente um número menor")
    elif palpite < num_sec:
      print ("tente um número maior")
  except ValueError:
      print ("digite um número válido")
```
