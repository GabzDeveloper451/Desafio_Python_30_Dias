```python
#Peça uma frase ao usuário, conte o número de vogais e substitua todos os espaços por sublinhados.

#cadastrando frase
frase_usuario = input("digite uma frase: ")

#definindo vogais
vogais = "aeiouAEIOU"
vog_ct = 0
for char in frase_usuario:
    if char in vogais:
        vog_ct += 1

#alterando espaço para _
frase_sublinhada = frase_usuario.replace(" ", "_")

#resultado final
print(f"Número de vogais na frase: {vog_ct}")
print(f"Frase com espaços substituídos por sublinhados: '{frase_sublinhada}'")
```
