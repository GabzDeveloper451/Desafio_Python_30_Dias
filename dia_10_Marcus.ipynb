```python
#Use um laço while para contar de 1 a 10 e imprimir apenas os números pares.

#definindo variável
a = 1

#definindo laço
while a <= 10:
  if a % 2==0:
    print(f"{a} é par e par não é imparosle")
  a+=1
```
