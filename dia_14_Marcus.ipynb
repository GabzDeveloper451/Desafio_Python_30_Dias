```python
#Importe o módulo math e calcule a área de um círculo com base no raio fornecido pelo usuário.

#importando biblioteca
import math as mt

#definindo variavel
while True:
    try:
        raio_str = input("digite o raio do circulo para sabermos a área do circulo: ")
        raio = float(raio_str)
        break
    except ValueError:
        print("Digite um número válido.")

#cálculo da área do circulo utilizando a biblioteca
area = mt.pi * raio**2
print(f"{area:.2f}")
```
