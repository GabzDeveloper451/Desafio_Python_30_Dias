```python
#Implemente um sistema de notas que classifique a pontuação do usuário (A, B, C, D, F).

#Cadastrando a variável

while True:
  try:
    nota = int(input("Digite sua nota: "))

  # definindo notas
    if 8 <= nota <= 10:
      print("Tirou um A (lá)")
      break
    elif 6 <= nota < 8:
      print("Tirou um B (sí)")
      break
    elif 3 <= nota < 6:
      print("Tirou um C (dó)")
      break
    elif 1 <= nota < 3:
      print("Tirou um D (ré)")
      break
    elif 0 <= nota < 1:
      print("Tirou um F (mi)")
      break
    else:
      print("A nota deve ser um número entre 0 e 10.")
      continue
  except ValueError:
      print("Digite um número inteiro entre 0 e 10.")
      continue

print("""Lembrando o seguinte:\n\nA= lá\nB= sí\nC= dó\nD= ré\nF= mi""")
```
