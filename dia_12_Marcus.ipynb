```python
#Definindo calculo da média

def calcular_media(lista_num):
    if not lista_num:
        return 0
    return sum(lista_num) / 5

# inserção dos números
print("insira 5 números: ")

numeros_digitados = []
for i in range(5):
    while True:
        try:
            numero = float(input(f"Digite o número {i+1}: "))
            numeros_digitados.append(numero)
            break
        except ValueError:
            print(" digite um número válido.")

media_digitados = calcular_media(numeros_digitados)
print(f"a média da lista ({numeros_digitados}) é: {media_digitados}")
```
