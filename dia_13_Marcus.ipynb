```python
#Entenda e demonstre a diferença entre variáveis locais e globais em uma função.

#definindo a variavel global
x= 3
print(f"a variável x={x} aqui é global")

#definindo a variavel local
def var_local():
    x=5
    y=2
    print(f"a variável x={x} aqui é local")
var_local()

#print da variável global
print(f"aqui se coloco x={x} ele reconhece como global")

#tentativa de printar local fora do "def"
print(f"aqui se coloco x={y} ele não reconhece como local")
```
