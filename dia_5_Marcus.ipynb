```python
#Crie um programa que verifique se um número é par ou ímpar.


while True:
  #definir input
  while True:
    num_dig=input("digite um numero inteiro: ")
    try:
      numero= int(num_dig)
      break
    except ValueError:
      print("digite um numero valido")

  #configurar saida
  if numero % 2 == 0:
   print("seu número é par")
  else:
   print("seu número é impar")



  nov_tent = input("quer tentar novamente (S ou N)? ").upper()
  if nov_tent == "S":
      continue
  elif nov_tent == "N":
      break
  else:
      print("Opção inválida. Por favor, digite 'S' para sim ou 'N' para não.")

```
